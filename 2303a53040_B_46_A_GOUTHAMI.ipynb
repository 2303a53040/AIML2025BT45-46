```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/car price prediction.csv')
print(data)
```

```python
cleaned_data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
```

```python
key_columns = ['year', 'selling_price', 'km_driven', 'fuel', 'transmission', 'owner']
cleaned_data = cleaned_data[key_columns]
```

```python
cleaned_data = cleaned_data.dropna()
```

```python
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
sns.scatterplot(data=cleaned_data, x='year', y='selling_price', alpha=0.6)
plt.title('Car Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(data=cleaned_data, x='km_driven', y='selling_price', alpha=0.6, color='orange')
plt.title('Car Price vs. Mileage (km_driven)')
plt.xlabel('Mileage (km_driven)')
plt.ylabel('Selling Price')
```

```python
plt.tight_layout()
plt.show()
```
